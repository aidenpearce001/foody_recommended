```python
link = list(dict.fromkeys(df["brand_link"]))
link = [i for i in link if "ha-noi" in i]
link[128].split('/')[-1]
len(link)
```

```python
cols = ["điểm chất lượng","điểm phục vụ","điểm không gian","điểm giá cả","điểm vị trí","địa điểm","thời gian mở","thời gian đóng","khoảng giá","Tổng bình luận","bình luyện tuyệt vời","bình luận tốt","bình luận trung bình","bình luận kém","Tên nhà hàng"] 
dframe = {cols[i]:[] for i in range(len(cols))}
```

```python
import numpy as np

for i,v in enumerate(link):
    print(i)
    content = requests.get(v)
    soup    = BeautifulSoup(content.text, "lxml")
    if soup.find("div", {"class": "errorpage"}) !=None:
        continue
    else:
        rating = (soup.find("div", {"class": "microsite-point-group"}).text).replace("\n"," ")
        rating = list(dict.fromkeys(rating.split(" ")))
        rating = [float(i) for i in rating if re.match(r"[0-9:]+", i)]
        print(rating)
        for _,k in enumerate(rating):
            if isinstance(k, float):
                dframe[cols[_]].append(k)
            else:
                dframe[cols[_]].append(None)
            
            
        position = soup.findAll('div',attrs={'class':'res-common-add'})
        pos = []
        for div in position:
            pos.append(div.text)
        dframe[cols[5]].append(re.sub(r'[\s]+', ' ', re.sub(r'[^ \w]', '', pos[0]).lower()))
        open_time = re.findall(r"[0-9]{1,2}:[0-9]{1,2}", re.sub(r'[^ \w:]', '', soup.find("div", {"class": "micro-timesopen"}).text).replace("\n"," "
                                                                                                                                           ).lower())
#         opent , closet = re.findall(r"[0-9]{1,2}:[0-9]{1,2}", re.sub(r'[^ \w:]', '', soup.find("div", {"class": "micro-timesopen"}).text).replace("\n"," "
#                                                                                                                                            ).lower())
        if len(open_time) > 2 :
            opent  = open_time[0]
            closet = open_time[-1]
        elif len(open_time) == 0:
            opent  = None
            closet = None
        else:
            opent = open_time[0]
            closet = open_time[1]
        dframe[cols[6]].append(opent)
        dframe[cols[7]].append(closet)
        if soup.find('span', attrs={'itemprop':'priceRange'}) : 
            dframe[cols[8]].append(soup.find('span', attrs={'itemprop':'priceRange'}).text.replace("\n",""))
        else: 
            dframe[cols[8]].append(None)
        col = 9
        
        if soup.find('div', attrs={'class':'ratings-boxes'}) :
            try:
                comments = re.sub(r'[^ \w]', '', soup.find('div', attrs={'class':'ratings-boxes'}).text.replace("\n","")).lower()
                for _ in [i for i in re.sub(r'[\s]+', ' ', comments).split("tiêu chí")[0].split(" ") if i.isdigit()]:
                    dframe[cols[col]].append(_)
                    col +=1
            except:
                for i in range(col):
                    dframe[cols[col]] = None
                    col +=1
                continue
        dframe[cols[14]].append(v.split('/')[-1])
        print(v.split('/')[-1])
# dframe
# pd.DataFrame(dframe)
```
